```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import matplotlib as mpl
import scienceplots

# change the background color to a specific hex color
plt.style.use(['science', 'dark_background'])
plt.rcParams['axes.facecolor'] = '#282a36'
plt.rcParams['figure.facecolor'] = '#282a36'
colorcycle = ['#8be9fd', '#ff79c6', '#50fa7b', '#bd93f9', '#ffb86c', '#ff5555', '#f1fa8c', '#6272a4']
plt.rcParams['axes.prop_cycle'] = mpl.cycler(color=colorcycle)

# change dpi
plt.rcParams['figure.dpi'] = 1024

# 4.29
# numerical integration
def integrand(u):
    return 1 / np.sqrt(1 - u**4)

integral = sp.integrate.quad(integrand, 0, 1)
tau = np.sqrt(8) * integral[0]
print(f"integral = {integral[0]} \n tau = {tau}")

# 4.33 
# plotting the function
r = m = g = 1
b_1 = 0.9 * r ; b_2 = 1.1 * r

def U(theta, b):
    return m * g * ((r + b) * np.cos(theta) + r * theta * np.sin(theta))

theta = np.linspace(-np.pi, np.pi, 1000)

# create subplots
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

# plot first subplot
axs[0].plot(theta, U(theta, b_1), label = "b = 0.9r")
axs[0].set_xlabel(r"$\theta$")
axs[0].set_ylabel(r"$U(\theta)$")
axs[0].set_xlim(-1.2, 1.2)
axs[0].set_ylim(1.85, 1.92)
axs[0].legend()

# plot second subplot
axs[1].plot(theta, U(theta, b_2), label = "b = 1.1r", color =colorcycle[1])
axs[1].set_xlabel(r"$\theta$")
axs[1].set_ylabel(r"$U(\theta)$")
axs[1].set_xlim(-2.5, 2.5)
axs[1].set_ylim(0, 2.5)
axs[1].legend()

# display the plots
plt.show()
```

```python
# 4.37
# Initial
g = R = M = 1
m_1 = 0.7 * M; m_2 = 0.8 * M
phi = np.linspace(-np.pi/4, 2*np.pi, 100)

def U(phi, m):
    return M * g * R * (1 - np.cos(phi)) - m * g * R * phi

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# first subplot
ax1.axhline(y=0, color='w', linewidth=0.5)
ax1.axvline(x=0, color='w', linewidth=0.5)
ax1.plot(phi, U(phi, m_1), label = "m = 0.7M")
ax1.set_xlabel(r"$\phi$")
ax1.set_ylabel(r"$U(\phi)$")
ax1.set_xlim(-np.pi/8, 3/2*np.pi)
ax1.set_ylim(-2, 0.5)
ax1.legend()
# drawiing x axis


# second subplot
ax2.axhline(y=0, color='w', linewidth=0.5)
ax2.axvline(x=0, color='w', linewidth=0.5)
ax2.plot(phi, U(phi, m_2), label = "m = 0.8M", color=colorcycle[2])
ax2.set_xlabel(r"$\phi$")
ax2.set_ylabel(r"$U(\phi)$")
ax2.set_xlim(-np.pi/8, 3/2*np.pi)
ax2.set_ylim(-2, 0.5)
ax2.legend()
# add arrow at the end of x-axis

# display the plots
plt.show()

# part d solving for phi
```
