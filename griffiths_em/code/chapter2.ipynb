```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import matplotlib as mpl
import scienceplots

# Science plot package + Dracula theme
plt.style.use(['science', 'dark_background'])
plt.rcParams['axes.facecolor'] = '#282a36'
plt.rcParams['figure.facecolor'] = '#282a36'
colorcycle = ['#8be9fd', '#ff79c6', '#50fa7b', '#bd93f9', '#ffb86c', '#ff5555', '#f1fa8c',
    '#6272a4']
plt.rcParams['axes.prop_cycle'] = mpl.cycler(color=colorcycle)
white = '#f8f8f2' # foreground

# change dpi
plt.rcParams['figure.dpi'] = 1024
```

```python
# 2.15
# Constants
con = sp.constants
e = 1; k = 1
a = 1; b = 2
r_ii = np.linspace(a, b, 1000)
r_iii = np.linspace(b, 5, 100)

# function from a to b
def E_in(r):
    return (k * (r - a)) / (e * r**2)

# function from b to infty
def E_out(r):
    return (k * (b - a)) / (e * r**2)

# plot
plt.figure().set_figwidth(5)
plt.plot([0, a], [0, 0], label='(i)')
plt.plot(r_ii, E_in(r_ii), label='(ii)')
plt.plot(r_iii, E_out(r_iii), label='(iii)')
plt.plot([b, b], [0, E_in(b)], '--')
plt.xlabel('r')
plt.ylabel(r'$|\mathbf{E}|$', rotation=0)
plt.xticks([0, a, b], [0, "a", 'b']); 
# plt.yticks(plt.yticks()[0], ['', '0', '', '', '', '', '', ''])
plt.yticks(plt.yticks()[0], ['' if i != 1 else '0' for i, _ in enumerate(plt.yticks()[0])])
plt.legend()
plt.show()
```

```python
# 2.16
# Constants
rho = e_o = 1
a = 1; b = 2
s_i = np.linspace(0, a, 100)
s_ii = np.linspace(a, b, 100)

# function from 0 to a
def E_i(s):
    return (rho * s) / (2 * e_o)

# function from a to b
def E_ii(s):
    return (rho * a**2) / (2 * e_o * s)

# plot
plt.figure().set_figwidth(5)
plt.plot(s_i, E_i(s_i), label='(i)')
plt.plot(s_ii, E_ii(s_ii), label='(ii)')
plt.plot([b, b], [E_ii(b), 0], '--')
plt.plot([b, 3 * a], [0, 0], label='(iii)')
plt.xlabel('s')
plt.ylabel(r'\(|\mathbf{E}|\)', rotation=0)
plt.xticks([0, a, b], [0, "a", 'b']);
plt.yticks(plt.yticks()[0], ['' if i != 1 else '0' for i, _ in enumerate(plt.yticks()[0])])
plt.legend()
plt.show()
```

```python
# 2.17
# Constants
rho = e_o = d = 1
y_i = np.linspace(-d, d, 1000)
y_ii = np.r_[np.linspace(-2*d, -d, 1000), np.linspace(d, 2*d, 1000)]
# y_ii = np.linspace(-4*d, -2*d, 1000)
# E from -2d to 2d
def E_i(y):
    return (rho * y) / (e_o)

# E outside |y| > 2d
def E_ii(y):
    return np.where(y > 0, (rho * d) / e_o, (-rho * d) / e_o)

# plot
plt.plot(y_ii, E_ii(y_ii), label=r'\( |y| > d \)', color=colorcycle[1])
plt.plot(y_i, E_i(y_i), label=r'\( |y| < d \)', color=colorcycle[0])
plt.xlabel('y')
plt.ylabel(r'$E$')
plt.xticks([-d, 0, d], ['-$d$', '0', '$d$']);
plt.yticks(plt.yticks()[0], ['' if val != 0 else '0' for i, val in enumerate(plt.yticks()[0])])
plt.legend()
plt.show()
```
